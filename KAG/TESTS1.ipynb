```python
import requests as rq
from rich import print as rprint

# Setting the URLs
urls = [
    "https://www.google.com",
    "https://www.yahoo.com",
    "https://www.meanbitches.com"
]

# Getting the user agent
uA = "Mozilla/5.0 (Linux; Android 11; Redmi Note 8 Pro) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Mobile Safari/537.36"

# Function to loop through URLs and make requests
def loop_requests():
    """Running the requests in a loop"""
    for url in urls:
        try:
            rez = rq.get(url, headers={"User-Agent": uA})
            rez.raise_for_status()  # Raise an exception for HTTP errors
            rprint(f"Request to {url} successful.")
            rprint(rez)
            rprint(rez.raw)
        except rq.exceptions.RequestException as e:
            rprint(f"Request to {url} failed: {e}")

loop_requests()
```

```python
import requests as rq
import logging
from rich import print as rprint

# Set up logging
logging.basicConfig(level=logging.INFO)

# Define URLs as a constant
URLS = [
    "https://www.google.com",
    "https://www.yahoo.com",
    "https://www.meanbitches.com"
]

# Define the user agent as a constant
USER_AGENT = "Mozilla/5.0 (Linux; Android 11; Redmi Note 8 Pro) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Mobile Safari/537.36"

def loop_requests() -> None:
    """Run requests in a loop"""
    for url in URLS:
        try:
            response = rq.get(url, headers={"User-Agent": USER_AGENT})
            response.raise_for_status()  # Raise an exception for HTTP errors
            logging.info(f"Request to {url} successful.")
            rprint(response)
            rprint(response.raw)
        except rq.exceptions.ConnectionError as e:
            logging.error(f"Connection error: {e}")
        except rq.exceptions.RequestException as e:
            logging.error(f"Request to {url} failed: {e}")

loop_requests()
```
